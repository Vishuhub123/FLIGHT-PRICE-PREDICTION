## 1 Delhi to Mumbai

```python
#connecting to web driver
driver=webdriver.Chrome('chromedriver.exe')

# for flights from delhi to bombay/mumbai
url='https://flight.yatra.com/air-search-ui/dom2/trigger?ADT=1&CHD=0&INF=0&class=Economy&destination=BOM&destinationCountry=IN&flexi=0&flight_depart_date=28%2F04%2F2022&hb=&noOfSegments=1&origin=DEL&originCountry=IN&type=O&unique=1650956310274&viewName=normal'
driver.get(url)
```

```python
name=[]
tdep=[]
tarr=[]
ttime=[]
tstop=[]
meals=[]
price=[]
city=[]
var=[]

# scrolling down
driver.execute_script("window.scrollBy(0,1000)") 
time.sleep(3)


#Scraping data of flight name 

nm = driver.find_elements_by_xpath('//div[@class="pr fl-info v-aligm-t text-center  pt-20 pl-24"] ')

try:
    for i in nm:
        name.append(i.text.split('\n')[0])
    
except NoSuchElementException:
    name.append('None')

#Scraping data of flight varient 

v = driver.find_elements_by_xpath('//div[@class="pr fl-info v-aligm-t text-center  pt-20 pl-24"] ')

try:
    for i in nm:
        var.append(i.text.split('\n')[-1])
    
except NoSuchElementException:
    var.append('None')

#Scraping data of departure time 

d = driver.find_elements_by_xpath('//div[@class="i-b pr"]')

try:
    for i in d:
        tdep.append(i.text)
    
except NoSuchElementException:
    tdep.append('None')
    
    
#Scraping data of arrival time 

a = driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')

try:
    for i in a:
        tarr.append(i.text.replace('\n+ 1 day',''))
    
except NoSuchElementException:
    tarr.append('None')
    
#Scraping data of Total time

tt = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')

try:
    for i in tt:
        ttime.append(i.text)
    
except NoSuchElementException:
    ttime.append('None')
    
#Scraping data of total stop 

ts = driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"] ')

try:
    for i in ts:
        tstop.append(i.text)
    
except NoSuchElementException:
    tstop.append('None')
    
#Scraping data of departure and arrival city name

ct = driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"] ')

try:
    for i in ct:
        city.append(i.text)
    
except NoSuchElementException:
    city.append('None')
    
#Scraping data of price

p = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')

try:
    for i in p:
        price.append(i.text.replace('₹',''))
    
except NoSuchElementException:
    price.append('None')
    


```

```python
name14=[]
tdep14=[]
tarr14=[]
ttime14=[]
tstop14=[]
meals14=[]
price14=[]
city14=[]
var14=[]

# scrolling down
driver.execute_script("window.scrollBy(0,1000)") 
time.sleep(3)


#Scraping data of flight name 

nm = driver.find_elements_by_xpath('//div[@class="pr fl-info v-aligm-t text-center  pt-20 pl-24"] ')

try:
    for i in nm:
        name14.append(i.text.split('\n')[0])
    
except NoSuchElementException:
    name14.append('None')

#Scraping data of flight varient 

v = driver.find_elements_by_xpath('//div[@class="pr fl-info v-aligm-t text-center  pt-20 pl-24"] ')

try:
    for i in nm:
        var14.append(i.text.split('\n')[-1])
    
except NoSuchElementException:
    var14.append('None')

#Scraping data of departure time 

d = driver.find_elements_by_xpath('//div[@class="i-b pr"]')

try:
    for i in d:
        tdep14.append(i.text)
    
except NoSuchElementException:
    tdep14.append('None')
    
    
#Scraping data of arrival time 

a = driver.find_elements_by_xpath('//p[@class="bold fs-15 mb-2 pr time"]')

try:
    for i in a:
        tarr14.append(i.text.replace('\n+ 1 day',''))
    
except NoSuchElementException:
    tarr14.append('None')
    
#Scraping data of Total time

tt = driver.find_elements_by_xpath('//p[@class="fs-12 bold du mb-2"]')

try:
    for i in tt:
        ttime14.append(i.text)
    
except NoSuchElementException:
    ttime14.append('None')
    
#Scraping data of total stop 

ts = driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"] ')

try:
    for i in ts:
        tstop14.append(i.text)
    
except NoSuchElementException:
    tstop14.append('None')
    
#Scraping data of departure and arrival city name

ct = driver.find_elements_by_xpath('//p[@class="fs-10 font-lightgrey no-wrap city ellipsis"] ')

try:
    for i in ct:
        city14.append(i.text)
    
except NoSuchElementException:
    city14.append('None')
    
#Scraping data of price

p = driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')

try:
    for i in p:
        price14.append(i.text.replace('₹',''))
    
except NoSuchElementException:
    price14.append('None')
    


departure14=city14[::2]
arr14=city14[1::2]
print(len(name14))
print(len(tdep14))
print(len(tarr14))
print(len(ttime14))
print(len(tstop14))
print(len(departure14))
print(len(arr14))
print(len(price14))
print(len(var14))

```
